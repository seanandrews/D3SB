# Controlled Tests

I want to generate a synthetic dataset where I know the true intensity profile as a controlled test of the D3SB inference approach.  I have done this with a similarity solution brightness profile

$I(r) = A \, \left( \frac{r}{r_c} \right)^{-\gamma} \exp{\left[-\left(\frac{r}{r_c}\right)^{2-\gamma}\right]}$

for a set of parameters $\{A, \, r_c, \, \gamma\}$ that denote an overall normalization, characteristic radius, and index.  If I re-cast this radial profile into projected angular coordinates, the azimuthally averaged brightness profile looks like this:

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

dpc = 140.
nr = 300
ftot = 0.1
rc  = 35./dpc
gam = 1.0
A = (2.-gam)*ftot/(2.*np.pi*rc**2)
r  = np.logspace(-2, 1, num=nr)
Ir = A * (r/rc)**(-gam) * np.exp(-(r/rc)**(2.-gam))

plt.axis([0.01, 5, 1e-9, 27])
plt.xscale('log')
plt.yscale('log')
plt.plot(r, Ir, '-')
```

In practice, this means I have generated a two-dimensional .fits image of this model that can be Fourier transformed and properly sampled.  Let's check that I didn't fuck this up:

```python
# - constants
AU = 1.496e13                 # - definition of an AU [cm]
  
# - parameter values
rin   = 0.1/dpc
incl = 0.
PA = 0.
  
# - radial grid
nx = 1001                 # - preferrably an odd number
ny = 1001                 # - best if this is square
rwid = 300.*AU            # - 1/2 width of image
xps = 2.*rwid/(nx-1)
yps = 2.*rwid/(ny-1)
xp_,yp_ = np.meshgrid(xps*(np.arange(nx)-(nx/2.-0.5)),yps*(np.arange(ny)-(ny/2.-0.5)))
ang = (270.-PA)*np.pi/180.
xp = np.cos(ang)*xp_-np.sin(ang)*yp_
yp = np.sin(ang)*xp_+np.cos(ang)*yp_
rr = np.sqrt(xp*xp+1./np.cos(incl*np.pi/180.)**2*yp*yp)/AU/dpc

# - create image
image = A*(rr/rc)**(-gam)*np.exp(-(rr/rc)**(2.-gam))
image[rr<rin] = 0.

rd = rr.flatten()
Ird = image.flatten()

plt.axis([0.01, 5, 1e-9, 27])
plt.xscale('log')
plt.yscale('log')

plt.plot(r, Ir, '-', rd, Ird, '.')
```

Ok excellent.  Now let's take a look at the synthetic visibility data:

```python
from get_vis import get_vis

data = get_vis('model.vis.fits')
u, v, dreal, dimag, dwgt = data

rho = np.sqrt(u**2+v**2)

plt.axis([0, 650, -0.1*np.amax(dreal), 1.1*np.amax(dreal)])
plt.plot(rho, dreal, '.', np.arange(650), np.zeros_like(np.arange(650)), '--')

dwgt = 1000.*dwgt
sigma = 1./np.sqrt(dwgt)
snr = dreal/sigma
print(np.mean(snr), np.std(snr))
```

Now let's see if we can (manually) reconstruct these profiles with the D3SB method:

```python
from d3sb_model import d3sb_model

# choose radial bins
nbins = 40
ba = np.logspace(-3, 1, num=nbins)
a = np.roll(ba, 1)
a[0] = rin
b = ba
bins = rin, b

# initial guess for weights
cb = 10.**(0.5*(np.log10(b)+np.log10(a)))
w0 = A*(cb/rc)**(-gam)*np.exp(-(cb/rc)**(2-gam))

Irings = np.zeros_like(r)
for i in range(nbins):
    Irings[(r >= a[i]) & (r <= b[i])] = w0[i]

uvsamples = u, v

# run the D3SB
mvis = d3sb_model(w0, uvsamples, bins)
print(mvis)

# Gaussian guess
sig = 0.75
nrm = 0.1/(2.*np.pi*sig**2)
w0 = nrm*np.exp(-0.5*(cb/sig)**2)

fig, ax = plt.subplots(1, 2, figsize=(12, 4))
ax[0].axis([0.01, 5, 1e-9, 27])
ax[0].set_xscale('log')
ax[0].set_yscale('log')
ax[0].plot(r, Ir, r, Irings,'r', cb, w0, 'b-')
ax[0].set_xlabel('angle [arcsec]')
ax[0].set_ylabel('intensity [Jy/arcsec2]')
ax[1].axis([0, 650, -0.1*np.amax(dreal), 1.1*np.amax(dreal)])
ax[1].plot(rho, dreal, '.', rho, mvis, '.', np.arange(650), np.zeros_like(np.arange(650)), '--')
ax[1].set_xlabel('angular frequency [klambda]')
ax[1].set_ylabel('real flux [Jy]')
```
