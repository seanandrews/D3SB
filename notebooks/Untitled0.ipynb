```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

dpc = 140.
nr = 500
ftot = 0.1
rc  = 35./dpc
gam = 1.0
A = (2.-gam)*ftot/(2.*np.pi*rc**2)
r  = np.logspace(-2, 1, num=nr)
Ir = A * (r/rc)**(-gam) * np.exp(-(r/rc)**(2.-gam))


# choose radial bins: b is the outer edge (passed)
# a (inner edge) and cb (center) are also defined for plotting purposes
nbins = 10
b = np.linspace(0.1, 2., num=nbins)
a = np.roll(b, 1)
rin = 0.01/140.
a[0] = rin
cb = 0.5*(a+b)

# true discretized amplitudes
w0 = A*(cb/rc)**(-gam)*np.exp(-(cb/rc)**(2-gam))

# initial guess based on Gaussian fit to fake visibilities
wg = np.ones_like(cb)
sig = 0.15
nrm = 0.1/(2.*np.pi*sig**2)
wg = nrm*np.exp(-0.5*(cb/sig)**2)
wg[wg <= 1e-4] = 1e-4


ndim = nbins+1
chain = np.load('saved_results.npy')
trace = chain[:, 10000:, :].reshape(-1, nbins)
print(trace.shape)

# set up a plot
fig, ax = plt.subplots(1, 2, figsize=(12, 4))
ax[0].axis([0.01, 10, 1e-5, 27])
ax[0].set_xscale('log')
ax[0].set_yscale('log')
ax[0].plot(r, Ir, 'black', cb, w0, 'ro', cb, wg, 'go')
ax[0].set_xlabel('angle [arcsec]')
ax[0].set_ylabel('intensity [Jy/arcsec2]')

ax[1].axis([0.0, 1.5, 1e-5, 27])
ax[1].set_yscale('log')
ax[1].set_xlabel('angle [arcsec]')
ax[1].set_ylabel('intensity [Jy/arcsec2]')

# draw randomly from posteriors and plot
for i in np.random.randint(len(trace), size=500):
    amps = trace[i][:]
    ax[0].plot(cb, 10.**amps, color="k", alpha=0.08)
    ax[1].plot(cb, 10.**amps, color="k", alpha=0.08)
    
ax[0].plot(r, Ir, 'black', cb, w0, 'ro', cb, wg, 'go')
ax[1].plot(r, Ir, 'black', cb, w0, 'ro', cb, wg, 'go')

optf = np.array([0.11553516, -0.91827446, -1.5817973, -2.09296416, -2.611059, -3.04614291, -3.48570132, -4.01786193, -4.24008815, -4.72330602])

ax[0].plot(cb, 10.**optf, 'bo')
ax[1].plot(cb, 10.**optf, 'bo')
```
