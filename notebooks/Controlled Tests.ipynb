# Controlled Tests

Ok, so in the "concepts" notebook I explained the basic premise of the "agnostic" discretized model of the surface brightness distribution.  Now I want to see if this actually will work in practice.  To start, I generated a fake dataset where I know the true intensity profile.  There is noise in the data, in that each visibility point comes attached with an uncertainty, but the underlying datapoints have not actually been perturbed from their true amplitudes.  I think of this as a (very) controlled test of the ability for "recovery" in the inference approach.    

The underlying model is a similarity solution brightness profile,

$I(r) = A \, \left( \frac{r}{r_c} \right)^{-\gamma} \exp{\left[-\left(\frac{r}{r_c}\right)^{2-\gamma}\right]}$

for a set of parameters $\{A, \, r_c, \, \gamma\}$ that denote an overall normalization, characteristic radius, and index.  This is perhaps a challenging first test, since it spans logarithmic scales in radius, but it should be vaguely representative of real disk behavior.  

First I define the "true" model parameters:

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

dpc = 140.
nr = 500
ftot = 0.1
rc  = 35./dpc
gam = 1.0
A = (2.-gam)*ftot/(2.*np.pi*rc**2)
r  = np.logspace(-2, 1, num=nr)
Ir = A * (r/rc)**(-gam) * np.exp(-(r/rc)**(2.-gam))
```

Separately, I generated a high resolution model image with this brightness profile, assuming a 60 degree inclination.  This was then Fourier transformed and sparsely sampled onto a set of visibility measurements (in this case, the CARMA 1.3mm GG Tau data from my recent paper; just what I had lying around).  The "noise" (visibility weights) is preserved from that dataset in this process.

I can retrieve that fake visibility data:

```python
from get_vis import get_vis

data = get_vis('model_faceon.vis.fits')
u, v, dreal, dimag, dwgt = data
rho = np.sqrt(u**2+v**2)
```

The success of this approach probably hinges (considerably) on an appropriate selection of the discretization; i.e., how we select the radial bins on which to define the model.  The optimized approach for doing this is not yet obvious to me.  Clearly we want to choose enough bins to capture variations in the brightness profile, but not too many that we can't constrain their amplitudes with the data (the computing time itself scales ~linearly with the number of bins we adopt; maybe that could be improved, but I'm not sure how).  To start, I'm experimenting with simple linear bins:  

```python
# choose radial bins: b is the outer edge (passed)
# a (inner edge) and cb (center) are also defined for plotting purposes
nbins = 15
b = np.linspace(0.1, 3., num=nbins)
a = np.roll(b, 1)
rin = 0.01/140.
a[0] = rin
cb = 0.5*(a+b)

# to be passed to model routine
bins = rin, b
```

Now I want to define the discretized amplitudes I *should* get if the code ran perfectly, written below as $\{w_{\rm true}\}$.  And I also have to define a starting "guess" to initialize the MCMC routine.  After some experimentation, it turns out that this is a crucially important step.  It seems that the best step is to run a standard optimization technique to get a good initial guess.  To do that, you need a sort of "initial, initial" guess, which I take from an elliptical Gaussian fit to the visibilities (with a fixed floor).

```python
# true discretized amplitudes
wtrue = A*(cb/rc)**(-gam)*np.exp(-(cb/rc)**(2-gam))
print(wtrue)

# initial, initial guess based on Gaussian fit to fake visibilities
sig = 0.15
nrm = 0.1/(2.*np.pi*sig**2)
wg = nrm*np.exp(-0.5*(cb/sig)**2)
wg[wg <= 1e-4] = 1e-4
print(wg)

# optimization 
from d3sb_model import d3sb_model
from lnprob import lnprob
from opt_func import opt_func
from scipy.optimize import minimize
opt = minimize(opt_func, np.log10(wg), args=(data, bins))
w0 = 10.**(opt.x)
print(w0)
```

Now I can check and see how this all goes together.  First I compute the visibilities for the discrete models (both "true" and "guess"), and compare them to the fake dataset through their $\chi^2$ values.  Then I plot up the comparisons: (top left) brightness profiles in log-log with black = true, gray = true but discretized, green = discretized guess; (top right) visibilities (unbinned) with error bars in blue, and corresponding models; (bottom left) same as top left but on a linear scale; and (bottom right) same as top right, but now showing just the fake visibility amplitudes (and not their corresponding errors) in blue, for clarity.

```python
# make a parameter array
#theta0 = np.append([incl0], [np.log10(w0)])
#thetag = np.append([inclg], [np.log10(wg)])
theta0 = np.log10(w0)
thetag = np.log10(wg)
thetat = np.log10(wtrue)

# run the D3SB to get visibility models
uvsamples = u, v
vis0 = d3sb_model(theta0, uvsamples, bins)
visg = d3sb_model(thetag, uvsamples, bins)
vist = d3sb_model(thetat, uvsamples, bins)

# compute and print the chi2 values
prob0 = lnprob(theta0, data, bins)
probg = lnprob(thetag, data, bins)
probt = lnprob(thetat, data, bins)
print(-2.*prob0, -2.*probg, -2.*probt)

# plot the results
fig, ax = plt.subplots(1, 2, figsize=(12, 4))
ax[0].axis([0.01, 10, 1e-6, 27])
ax[0].set_xscale('log')
ax[0].set_yscale('log')
ax[0].plot(r, Ir, 'k', cb, w0, 'ro', cb, wg, 'go')
ax[0].set_xlabel('angle [arcsec]')
ax[0].set_ylabel('intensity [Jy/arcsec2]')
ax[1].axis([0, 650, -0.5, 0.6])  
ax[1].plot(rho, visg, 'g.', rho, vis0, 'r.', np.arange(650), np.zeros_like(np.arange(650)), 'k--')
ax[1].errorbar(rho, dreal, yerr=np.sqrt(1./dwgt), ecolor='b', fmt='.r', alpha=0.003)
ax[1].set_xlabel('angular frequency [klambda]')
ax[1].set_ylabel('real flux [Jy]')

fig, ax = plt.subplots(1, 2, figsize=(12, 4))
ax[0].axis([0.0, 1.5, 1e-6, 27.])
ax[0].set_yscale('log')
ax[0].plot(r, Ir, 'black', cb, w0, 'ro', cb, wg, 'go')
ax[0].set_xlabel('angle [arcsec]')
ax[0].set_ylabel('intensity [Jy/arcsec2]')
ax[1].axis([0, 650, -0.1*np.amax(dreal), 1.1*np.amax(dreal)])
ax[1].plot(rho, visg, 'g.', rho, vist, 'ko', rho, vis0, 'r.', np.arange(650), np.zeros_like(np.arange(650)), 'k--')
ax[1].set_xlabel('angular frequency [klambda]')
ax[1].set_ylabel('real flux [Jy]')
```

Now we can run an MCMC simulation with $emcee$ to see if we can recover the "true" surface brightness distribution:

```python
import emcee
from lnprob import lnprob

# set up walkers
ndim, nwalkers, nthreads = nbins, 80, 8
p0 = theta0
p0 = [p0 + np.random.uniform(-2, 2, ndim) for i in range(nwalkers)]

# initialize sampler
sampler = emcee.EnsembleSampler(nwalkers, ndim, lnprob, threads=nthreads, args=[data, bins])

# run emcee
%time sampler.run_mcmc(p0, 100)
```

Now we want to take a look at those results compared to the "true" situation:

```python
from astropy.io import ascii
from astropy.table import Table

# reshape the chain output
trace = sampler.chain[:, 50:, :].reshape(-1, ndim)

# set up a plot
fig, ax = plt.subplots(1, 2, figsize=(12, 4))
ax[0].axis([0.01, 10, 1e-9, 27])
ax[0].set_xscale('log')
ax[0].set_yscale('log')
ax[0].plot(r, Ir, 'black', cb, w0, 'ro', cb, wg, 'go')
ax[0].set_xlabel('angle [arcsec]')
ax[0].set_ylabel('intensity [Jy/arcsec2]')

ax[1].axis([0.0, 3, 1e-9, 27])
ax[1].set_yscale('log')
ax[1].set_xlabel('angle [arcsec]')
ax[1].set_ylabel('intensity [Jy/arcsec2]')

# draw randomly from posteriors and plot
for i in np.random.randint(len(trace), size=500):
    amps = trace[i][:]
    ax[0].plot(cb, 10.**amps, color="y", alpha=0.1)
    ax[1].plot(cb, 10.**amps, color="y", alpha=0.1)
    
ax[0].plot(r, Ir, 'black', cb, w0, 'ro', cb, wg, 'go')
ax[1].plot(r, Ir, 'black', cb, w0, 'ro', cb, wg, 'go')
```
