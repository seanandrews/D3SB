# Notes on Surface Brightness Decomposition

Interferometers measure the Fourier transform (FT) of the sky brightness.  We start here with the assumption that the surface brightness of the source is azimuthally symmetric (purely radial), $I(r)$, where $r = \sqrt{x^2+y^2}$.  The FT of a radial function is a Hankel transform,

$V(\rho) = {\rm FT}\{I(r)\} = \int_0^{\infty} I(r) \, J_0(2\pi \, \rho \, r) \, r \, dr$,

where $V$ denotes the complex visibilities, $J_0$ is the zeroth order Bessel function of the first kind and $\rho = \sqrt{u^2+v^2}$ is a radial coordinate in the Fourier domain.  For most functional forms of $I(r)$, the corresponding $V(\rho)$ must be determined numerically, either by integrating through the Hankel transform or (more commonly) employing a fast Fourier transform (FFT) algorithm.  In the context of modeling $V(\rho)$, this requires that one knows the parametric form for $I(r)$: but usually that form is not known a priori.

I am aiming to develop a "model agnostic" approach to fitting the visibility data.  Going back to the general Hankel transform above, we can decompose $I(r)$ into an infinite sequence of weighted $\delta$-function rings.  Each such ring has a visibility function

$V_{ring}(\rho) = \int_0^{\infty} w(r^{\prime}) \, \delta(r-r^{\prime}) \, J_0(2\pi \, \rho \, r) \, r \, dr = 2\pi \, r^{\prime} \, w(r^{\prime}) \, J_0(2\pi \, \rho \, r^{\prime})$.

Therefore, by the linearity of the FT, the total visibility function of any arbitrary brightness function can be determined by integrating over the ensemble of rings,

$V(\rho) = \int_0^{\infty} V_{ring}(\rho) \, dr^{\prime} = 2\pi \, \int w(r^{\prime}) \, J_0(2\pi \, \rho \, r^{\prime}) \, r^{\prime} \, dr^{\prime}$.

In essence, $w(r)$ is an infinite sequence of "weights" that traces out the arbitrary continuous functional form of $I(r)$.  

The above "decomposition" is, of course, nothing more than a semantic re-writing of the original Hankel transform.  But, it serves to highlight the mechanics of our "leap" to a more practical discretized decomposition.  Suppose instead that I break the weighting function $w$ into larger, non-infinitesimal chunks.  In essence, rather than assuming $\delta$-functions as the decomposition basis functions, I can approximate the Hankel transform as a discrete weighted sum of uniform rings with finite widths.  

A single such ring can be described by the radial boxcar function,

$\Pi_{ab}(r) = w \left[ H_b(r) - H_a(r) \right]$,

where $H_{r^{\prime}}(r)$ is the Heaviside step function, equal to unity when $r \le r^{\prime}$.  In this case, $\Pi_{ab}(r) = w$ when $a \le r \le b$ and zero otherwise.  The Fourier (Hankel) transform of the Heaviside function is

${\rm FT}\{H_a(r)\} = \int_0^{\infty} H_a(r) \, J_0(2\pi \, r \, \rho) \, r \, dr = \frac{\pi a^2}{\pi a \rho} \, J_1(2\pi \, a \, \rho) = \frac{a}{\rho} \, J_1(2\pi \, a \, \rho)$

where $J_1$ is the first order Bessel function of the first kind.  Then, again by the linearity of the FT, a single ring between radii of $a$ and $b$ with intensity $w$ has a FT

$V_{ring}(\rho) = {\rm FT}\{\Pi_{ab}(r)\} = w \left[{\rm FT}\{H_b(r)\} - {\rm FT}\{H_a(r)\}\right] = \frac{w}{\rho} \left[ b \, J_1(2\pi \, b \, \rho) - a \, J_1(2\pi \, a \, \rho) \right]$.

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.special as sc
%matplotlib inline

# choose the ring parameters in radial space
# a and b are defined here in arcseconds
w = 0.1
a = 2.5
b = 2.85

# generate a synthetic radial ring profile
nr = 401
r = 0.025*np.arange(nr)
ring = np.zeros_like(r)
ring[(r >= a) & (r <= b)] = w

# convert radial scales into proper units (radians)
aa = a/206264.806427
bb = b/206264.806427

# define a radial frequency coordinate (in wavelength units)
nrho = 700
rho = 1e3*np.linspace(1, nrho, num=nrho)

# FT the ring (amplitude only; this is symmetric radially)
# note that you need the " --> radians normalization factor if a, b, and w are defined wrt "
Vrho = 206264.806427*(w/rho)*(b*sc.jn(1, 2*np.pi*bb*rho)-a*sc.jn(1, 2*np.pi*aa*rho))

# plot the intensity profile and its FT
fig, ax = plt.subplots(1, 2, figsize=(12, 4))
ax[0].axis([0, 10, -0.1*w, 1.1*w])
ax[0].plot(r, ring, r, np.zeros(nr))
ax[0].set_xlabel('angle [arcsec]')
ax[0].set_ylabel('intensity [Jy/arcsec2]')
ax[1].axis([0, 500, np.amin(Vrho)-0.1*np.amax(Vrho), 1.1*np.amax(Vrho)])
ax[1].plot(rho*1e-3, Vrho, rho*1e-3, np.zeros(nrho))
ax[1].set_xlabel('angular frequency [klambda]')
ax[1].set_ylabel('real flux [Jy]')
```

Following the semantic example for a continuous basis (the $\delta$-functions), an arbitrary $I(r)$ can be decomposed into a discrete sum (again by the linearity of the FT) of these uniform-intensity rings

$V(\rho) = \sum_i V_{ring}^i = \sum_i \frac{w_i}{\rho} \left[b_i \, J_1(2\pi \, b_i \, \rho) - a_i \, J_1(2\pi \, a_i \, \rho) \right]$,

for a finite sequence of weights $\{w_i\}$ and ring boundaries \{$a_i$, $b_i$\}.

As an example, consider the simple analytic case of a Gaussian function with $\mu = 0$ and an arbitrary width $\sigma$,

$I(r) = \exp \left[ -\frac{1}{2}\left(\frac{r}{\sigma}\right)^2\right]$,

which can be artificially discretized on a linear sequence of radial bins with weights

$w_i = \exp\left[-\frac{1}{2} \left( \frac{a_i+b_i}{2 \sigma} \right)^2 \right]$.

The analytic FT is calculated as

$V(\rho) = {\rm FT}\{I(r)\} = 2\pi \, \sigma^2 \exp\left[-2\pi^2 \, \rho^2 \, \sigma^2\right]$,

and its discretized equivalent is

$V(\rho) \approx \sum_i \frac{1}{\rho} \, \exp\left[-\frac{1}{2} \left( \frac{a_i+b_i}{2 \sigma} \right)^2 \right] \left[b_i \, J_1(2\pi \, b_i \, \rho) - a_i \, J_1(2\pi \, a_i \, \rho) \right]$.



```python
# make a continuous Gaussian intensity profile
sig = 0.5
Inu = np.exp(-0.5*r**2/sig**2)

# assign radial bins
nbins = 45
dr = 0.25
b = dr+dr*np.arange(nbins)
a = b-dr

# assign weights
w = np.exp(-0.5*((a+b)/(2*sig))**2)

# make a discretized version
Irings = np.zeros_like(r)
for i in range(nbins):
    Irings[(r >= a[i]) & (r <= b[i])] = w[i]
    
# analytic (continuous) FT
Vana = 2*np.pi*sig**2*np.exp(-2*(np.pi*rho*sig/206264.806427)**2)

# discretized FT
Vdis = np.zeros_like(rho)
for i in range(nbins):
    Vdis += 206264.806427*(w[i]/rho)*(b[i]*sc.jn(1, 2*np.pi*b[i]*rho/206264.806427)-a[i]*sc.jn(1, 2*np.pi*a[i]*rho/206264.806427))

# plot the intensity profile and its FT
fig, ax = plt.subplots(1, 2, figsize=(12, 4))
ax[0].axis([0, 6, -0.1*np.amax(Inu), 1.1*np.amax(Inu)])
ax[0].plot(r, Irings, '-', r, np.zeros(nr), r, Inu, '--')
ax[0].set_xlabel('angle [arcsec]')
ax[0].set_ylabel('intensity [Jy/arcsec2]')
ax[1].axis([0, 500, np.amin(Vana)-0.1*np.amax(Vana), 1.1*np.amax(Vana)])
ax[1].plot(rho*1e-3, Vdis, '-', rho*1e-3, np.zeros(nrho), rho*1e-3, Vana, '--')
ax[1].set_xlabel('angular frequency [klambda]')
ax[1].set_ylabel('real flux [Jy]')    
```
