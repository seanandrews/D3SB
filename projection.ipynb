# Projection

To handle the on-sky projection of a disk oriented with a given inclination angle $i$ (relative to our line-of-sight) and major axis position angle $\varphi$ (relative to north), we can simply impose a coordinate change in the Fourier plane.  Since the FT relations for Gaussian brightness distributions are straightforward and analytic, it is helpful to use them when illustrating a concrete example for how to do this coordinate change properly.   

First create an image plane model for a symmetric (circular) Gaussian, $g_0(x, y)$,

$g_0(x, y) = a \, \exp{\left[-\frac{\sqrt{x^2+y^2}}{2 \sigma^2}\right]}$.

The FT of that distribution is another Gaussian, with amplitude $A = 2\pi \, \sigma^2 \, a$ and width $\Sigma = 1/2\pi\,\sigma$,

$G_0(u, v) = {\rm FT}\{g_0(x, y)\} = A \, \exp{\left[-\frac{\sqrt{u^2+v^2}}{2 \Sigma^2}\right]} = 2\pi \, \sigma^2 \, a \, \exp{\left[-2\pi^2 \, \sigma^2 \, (u^2 + v^2)\right]}$. 

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
%matplotlib inline

# image plane coordinates
nx, dx, ny, dy = 1001, 0.02, 1001, 0.02
x = -10 + dx*np.arange(nx)
y = -10 + dy*np.arange(ny)

# image plane Gaussian parameters
sig = 2.0
nrm = 0.1

# image plane Gaussian
xp, yp = np.meshgrid(x, y)
r = np.sqrt(xp**2 + yp**2)
g0 = nrm*np.exp(-0.5*r**2/sig**2)

# Fourier plane coordinates
nu, du, nv, dv = 1001, 0.25, 1001, 0.25
u = -125 + du*np.arange(nu)
v = -125 + dv*np.arange(nv)

# Fourier plane Gaussian: FT(g0)
up, vp = np.meshgrid(u, v)
rho = np.sqrt(up**2 + vp**2)
rhol = 1e3*rho/206264.806427
G0 = 2.*np.pi*sig**2*nrm*np.exp(-2.*np.pi**2*sig**2*rhol**2)

fig, ax = plt.subplots(1, 2, figsize=(12, 4))
ax[0].imshow(g0, extent=[10, -10, -10, 10], cmap='bone_r', vmax=0.1, vmin=0.0)
ax[0].set_xlabel('relative RA [arcsec]')
ax[0].set_ylabel('relative DEC [arcsec]')
ax[1].imshow(G0, extent=[125, -125, -125, 125], cmap='summer_r')
ax[1].set_xlabel('u [klambda]')
ax[1].set_ylabel('v [klambda]')
```

Now, we can impart a projection into the plane of the sky (an inclination) in several ways.  In the Gaussian case, this could be done by assigning different $\sigma$ values for the $x$ and $y$ dimensions (e.g., in Isella et al. 2013).  For a geometry where "face-on" is represented by $i = 0^{\circ}$, and where we tilt the disk along a fixed axis aligned with declination (so that the default position angle of the major axis is $\varphi = 0^{\circ}$), then this would mean that $\sigma_x = \sigma_y \, \cos{i}$.  It makes more sense to think of this as a coordinate change, $x^{\prime} = x \,/\cos{i}$.

In that case, the image plane model for an inclined Gaussian, $g^{\prime}(x, y)$, is just

$g^{\prime}(x^{\prime}, y) = a \, \exp{\left[-\frac{\sqrt{x^{{\prime} 2}+y^2}}{2 \sigma^2}\right]} = a \, \exp{\left[-\left(\frac{x^2}{2(\cos{i})^2 \, \sigma^2}\right) - \left(\frac{y^2}{2\sigma^2}\right) \right]}$.

The FT of that distribution is another Gaussian, with amplitude $A = 2\pi \, \sigma^2 \, a$ and width $\Sigma = 1/2\pi\,\sigma$,

$G_0(u, v) = {\rm FT}\{g_0(x, y)\} = A \, \exp{\left[-\frac{\sqrt{u^2+v^2}}{2 \Sigma^2}\right]} = 2\pi \, \sigma^2 \, a \, \exp{\left[-2\pi^2 \, \sigma^2 \, (u^2 + v^2)\right]}$. 

```python
incl = 60.*np.pi/180.
xxp = xp/np.cos(incl)
r = np.sqrt(xxp**2+yp**2)
g0_i = nrm*np.exp(-0.5*r**2/sig**2)

uup = up*np.cos(incl)
rho = np.sqrt(uup**2 + vp**2)
rhol = 1e3*rho/206264.806427
G0_i = 2.*np.pi*np.abs(np.cos(incl))*sig**2*nrm*np.exp(-2.*np.pi**2*sig**2*rhol**2)

fig, ax = plt.subplots(1, 2, figsize=(12, 4))
ax[0].imshow(g0_i, extent=[10, -10, -10, 10], cmap='bone_r', vmax=0.1, vmin=0.0)
ax[0].set_xlabel('relative RA [arcsec]')
ax[0].set_ylabel('relative DEC [arcsec]')
ax[1].imshow(G0_i, extent=[125, -125, -125, 125], cmap='summer_r')
ax[1].set_xlabel('u [klambda]')
ax[1].set_ylabel('v [klambda]')
```

```python
phi = 50.*np.pi/180.
rxp = (xp*np.cos(phi)-yp*np.sin(phi))/np.cos(incl)
ryp = xp*np.sin(phi)+yp*np.cos(phi)
rr = np.sqrt(rxp**2+ryp**2)
g0_ir = nrm*np.exp(-0.5*rr**2/sig**2)

rup = (up*np.cos(phi)-vp*np.sin(phi))*np.cos(incl)
rvp = up*np.sin(phi)+vp*np.cos(phi)
rho = np.sqrt(rup**2 + rvp**2)
rhol = 1e3*rho/206264.806427
G0_i = 2.*np.pi*np.abs(np.cos(incl))*sig**2*nrm*np.exp(-2.*np.pi**2*sig**2*rhol**2)

fig, ax = plt.subplots(1, 2, figsize=(12, 4))
ax[0].imshow(g0_ir, extent=[10, -10, -10, 10], cmap='bone_r', vmax=0.1, vmin=0.0)
ax[0].set_xlabel('relative RA [arcsec]')
ax[0].set_ylabel('relative DEC [arcsec]')
ax[1].imshow(G0_i, extent=[125, -125, -125, 125], cmap='summer_r')
ax[1].set_xlabel('u [klambda]')
ax[1].set_ylabel('v [klambda]')
```
